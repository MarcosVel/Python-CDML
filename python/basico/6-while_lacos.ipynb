# Laços while

A instrução <code>while</code> no Python é uma das maneiras mais gerais de executar a iteração. Uma instrução <code>while</code> executará repetidamente uma única instrução ou grupo de instruções, desde que a condição seja verdadeira. O motivo pelo qual é chamado de 'loop' é porque as instruções de código são repetidas repetidamente até que a condição não seja mais atendida.


O formato geral de um loop while é:

    while teste:
        código1
    else:
        código2

Vejamos alguns loops simples com <code>while</code>.

```python
x = 0

while x < 10:
    print('x é: ',x)
    print(' x ainda é menor que 10, adicionando 1 para x =======')
    x+=1
```

Observe quantas vezes as instruções de impressão ocorreram e como o loop <code>while</code> continuou até que a condição True fosse atendida, que ocorreu uma vez x==10. É importante observar que, uma vez que isso ocorreu, o código parou. Vamos ver como podemos adicionar uma instrução <code>else</code>:

```python
x = 50

while x < 10:
    print('x é: ',x)
    print(' x ainda é inferior a 10, adicionando 1 para x')
    x+=1 # x = x + 1   
else:
    print('Tudo feito!')
```

# break, continue, pass

Podemos usar as instruções <code>break</code>, <code>continue</code> e <code>pass</code> em nossos loops para adicionar funcionalidade adicional para vários casos. As três instruções são definidas por:

    break: Interrompe o loop fechado mais próximo atual.
    continue: Vai para o topo do loop mais próximo.
    pass: Não faz nada.
    
    
Pensando nas instruções <code>break</code> e <code>continue</code>, o formato geral do loop <code>while</code> fica assim:

    while teste: 
        codigo1
        if teste1: 
            break
        if teste2: 
            continue 
    else:

As instruções <code>break</code> e <code>continue</code> podem aparecer em qualquer lugar dentro do corpo do loop, mas geralmente as colocaremos ainda mais aninhadas em conjunto com uma instrução <code>if</code> para executar um ação com base em alguma condição.

Vamos em frente e veja alguns exemplos!

```python
x = 0

while x < 10:
    print('x é: ',x)
    print(' x ainda é menor que 10, adicionando 1 a x')
    x+=1
    if x==3:
        print('x==3')
    else:
        print('continuando...')
        continue
```

Observe como temos uma declaração impressa quando x == 3 e uma continuação sendo impressa à medida que continuamos pelo loop while externo. Vamos fazer uma pausa uma vez x == 3 e ver se o resultado faz sentido:

```python
x = 0

while x < 10:
    print('x é: ',x)
    print(' x ainda é menor que 10, adicionando 1 a x')
    x+=1
    if x==3:
        print('Quebrando porque x==3')
        break
    else:
        print('continuando...')
        continue
```

Observe como a outra declaração <code>else</code> não foi alcançada e a continuação nunca foi impressa!

Após esses exemplos breves, porém simples, você deve se sentir confortável usando as instruções <code>while</code> no seu código.

**Uma palavra de cautela, no entanto! É possível criar um loop infinitamente em execução com as instruções <code>while</code>. Por exemplo:**

```python
# NÃO EXECUTE ESTE CÓDIGO !!!!
while True:
    print("Estou preso em um loop infinito!")
```
