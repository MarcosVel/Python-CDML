# Plots de regresses

O Seaborn possui muitas ferramentas integradas para plots de regresso, no entanto, no discutiremos a regresso,  mas apenas cobriremos a funo **lmplot()**.

**lmplot** permite que voc exiba modelos lineares, mas tambm permite que voc divida esses grficos com base em recursos, alm de colorir a matiz de cores com base nos recursos.

https://pt.wikipedia.org/wiki/Regresso_linear




A regresso, em geral, tem como objectivo tratar de um valor que no se consegue estimar inicialmente.


Vamos explorar como isso funciona:


```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## lmplot()

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

### Trabalhando com marcadores

lmplot kwargs so passados atravs do **regplot**, que  uma forma mais geral de lmplot (). O regplot possui um parmetro scatter_kws  passado para plt.scatter. Ento, voc pode querer definir o parmetro "s" nesse dicionrio, o que corresponde ao tamanho dos marcadores. Em outras palavras, voc acaba passando um dicionrio com os argumentos base do matplotlib, neste caso, s para o tamanho do grfico de disperso. Em geral, voc provavelmente no vai se lembrar disso sempre, porm, consulte sempre que achar necessrio.

```python
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

```python
help(sns.lmplot)
```

## Usando grades

Podemos adicionar mais separao varivel atravs de colunas e linhas com o uso de uma grade. Basta indicar isso com os argumentos col ou row:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

## Aspecto e tamanho

As figuras de Seaborn podem ter seu tamanho e aspect ajustados com os parmetros **size** e **aspecto**:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,size=8)
```
