# PairGrids

Os PairGrids são tipos gerais de gráficos que permitem mapear tipos de plotagem diferentes para linhas e colunas de um grid, isso ajuda você a criar plots similares separadas por categoias.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
iris.head()
```

iris.png

## PairGrid

Pairgrid é um plot de grade para traçar relacionamentos entre pares de um conjunto de dados.

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
import numpy as np; 
np.random.seed(10)
import seaborn as sns; 
sns.set(color_codes=True)
mean, cov = [0, 2], [(1, .5), (.5, 1)]
x, y = np.random.multivariate_normal(mean, cov, size=50).T
sns.kdeplot(x)
sns.kdeplot(y)
```

```python
# Altera os tipos de plots na diagonal, parte superior e inferior.
g = sns.PairGrid(iris)
g.map_diag(plt.hist)  # diagonal 
g.map_upper(plt.scatter)  # diagonal superior
g.map_lower(sns.kdeplot) # diagonal inferior
```

## pairplot

Pairplot é uma versão mais simples do PairGrid (você usará com bastante frequência)

```python
sns.pairplot(iris)
```

```python
#hue é matiz - Dados variáveis para mapear aspectos de plotagem para cores diferentes.
sns.pairplot(iris,hue='species',palette='Reds')
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

## FacetGrid

FacetGrid é a maneira geral de criar plots de grades com base em um recurso:

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

```python
# Só a grade
g = sns.FacetGrid(tips, col="time", row="smoker")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="day")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="sex")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Observe como os argumentos vêm após a chamada do plt.scatter
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Observe como os argumentos vêm após a chamada do plt.scatter
g = g.map(plt.scatter, "total_bill", "tip", "size").add_legend()
```

## JointGrid

JointGrid é a versão geral para grades tipo jointplot (), para um exemplo rápido:

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```

```python
g = sns.JointGrid(x="total_bill", y="size", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```

```python
g = sns.JointGrid(x="tip", y="total_bill", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```
